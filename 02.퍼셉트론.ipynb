1. 입력 신호가 뉴런에 보내질 때는 각각 고유한 **가중치**가 곱해진다.  
2. 뉴런에서 보내온 신호의 총합이 정해진 한계를 넘어설때만 1을 출력한다.(뉴런이 활성화한다.)  
3. 그 한계를 **임계값**이라고한다.  
4. 퍼셉트론은 복수의 입력 신호 각각에 고유한 가중치를 부여한다.   
5. 가중치는 각 신호가 결과에 주는 영향력을 조절하는 요소로 작용한다.
6. 가중치가 클수록 해당 신호가 그만큼 더 중요함을 뜻한다. 

```python
def AND(x1,x2):
    w1,w2,theta=0.5,0.5,0.7
    tmp=x1*w1+x2*w2
    if tmp<=theta:
        return 0
    elif tmp>theta:
        return 1
```

```python
AND(0,0)
```

```python
AND(1,0)
```

```python
AND(0,1)
```

```python
AND(1,1)
```

weight와 bias를 도입한 AND 게이트  
- w1,w2는 각 입력 신호가 결과에 주는 영향력(중요도)을 조절하는 매개변수
- bias : 뉴런이 얼마나 쉽게 활성화(결과로 1을 출력) 하느냐를 조정하는 매개변수
- 이 모두 문맥에 따라 가중치 라고 부르기도 한다.

```python
def AND(x1,x2):
    x=np.array([x1,x2])
    w=np.array([0.5,0.5])
    b=-0.7
    tmp=np.sum(w*x)+b
    if tmp<=0:
        return 0
    else:
        return 1
```

```python
def NAND(x1,x2):
    x=np.array([x1,x2])
    w=np.array([-0.5,-0.5])
    b=0.7
    tmp=np.sum(w*x)+b
    if tmp<=0:
        return 0
    else:
        return 1

def OR(x1,x2):
    x=np.array([x1,x2])
    w=np.array([0.5,0.5]) # AND와는 가중치(w와b)만 다르다.
    b=-0.2
    tmp=np.sum(w*x)+b
    if tmp<=0:
        return 0
    else:
        return 1
    
```

XOR  
- 직선 하나로는 XOR의 1과 0을 나눌 수 없다. 하지만 '직선'이라는 제약을 없애면 가능하다.
- 따라서 곡선의 영역을 비선형 영역, 직선의 영역을 선형영역이라고 한다.
- 그렇기 때문에 퍼셉트론으로는 XOR 게이트를 표현할 수 없다. 
1. 퍼셉트론의 한계를 정확히 말하면 "단층 퍼셉트론"으로는 XOR 게이트를 표현할 수 없다. or "단층 퍼셉트론으로는 비선형 영역을 분리할 수 없다"
2. 퍼셉트론을 조합하여, 즉 층을 쌓아서 XOR 게이트를 구현한다. 

```python
def XOR(x1,x2):
    s1=NAND(x1,x2)
    s2=OR(x1,x2)
    y=AND(s1,s2)
    return y
```

```python
XOR(0,0)
```

```python
XOR(1,0)
```

```python
XOR(0,1)
```

```python
XOR(1,1)
```


- 따라서 퍼셉트론은 지금까지 본 AND, OR 퍼셉트론과 형태가 다르다.
- 실제로 AND, OR가 단층 퍼셉트론인데 반해, XOR은 2층 퍼셉트론이다.
- 이러한 층이 여러개인 퍼셉트론을 다층 퍼셉트론이라고한다. 
1. 0층의 두 뉴런이 입력 신호를 받아 1층의 뉴런으로 신호를 보낸다.
2. 1층의 뉴런이 2층의 뉴런으로 신호를 보내고, 2층의 뉴런은 y를 출력한다.

- 처음에는 AND와 OR 게이트, 그 다음에는 반가산기와 전가산기, 그 다음에는 산술 논리 연산 장치(ALU), 그 다음에는 CPU 이다.

### 정리
1. 퍼셉트론은 입출력을 갖춘 알고리즘이다. 입력을 주면 정해진 규칙에 따른 값을 출력한다.
2. 퍼셉트론에서는 '가중치'와 '편향'을 매개변수로 설정한다.
3. 퍼셉트론으로 AND, OR 게이트 등의 논리 회로를 표현할 수 없다.
4. 2층 퍼셉트론을 이용하면 XOR 게이트를 표현할 수 있다.
5. 단층 퍼셉트론은 직선형 영역만 표현할 수 잇고, 다층 퍼셉트론은 비선형 영역도 표현할 수 있다.
6. 다층 퍼셉트론은 (이론상) 컴퓨터를 표현할 수 있다.
